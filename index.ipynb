### Plot Log Transformed Versions of Features

For each feature that you identified as a good candidate for log transformation, plot the feature vs. `SalePrice` as well as the log transformed feature vs. log transformed `SalePrice`.

```python
# Your code here
# List of features identified for log transformation
features_to_transform = ['LotArea', 'MasVnrArea', 'GrLivArea']

# Set up subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))
for i, feature in enumerate(features_to_transform):
    ax_original = axes[i][0]
    ax_log = axes[i][1]
    
    # Plot original feature vs. SalePrice
    ax_original.scatter(ames[feature], y, alpha=0.5, color='blue')
    ax_original.set_xlabel(feature)
    ax_original.set_ylabel('SalePrice')
    ax_original.set_title(f'{feature} vs. SalePrice')
    
    
    log_feature = np.log(ames[feature] + 1) 
    log_y = np.log(y + 1)  
    
    # Plot log-transformed feature vs. log-transformed SalePrice
    ax_log.scatter(log_feature, log_y, alpha=0.5, color='red')
    ax_log.set_xlabel(f'Log({feature})')
    ax_log.set_ylabel('Log(SalePrice)')
    ax_log.set_title(f'Log({feature}) vs. Log(SalePrice)')

fig.tight_layout()
plt.show()
```

## Build a Model with Log-Transformed Features and Target

### Data Preparation

Choose up to 3 of the features you investigated, and set up an X dataframe containing the log-transformed versions of these features as well as a y series containing the log-transformed version of the target.

---

<details>
    <summary style="cursor: pointer"><b>Hint (click to reveal)</b></summary>

If you are planning log transform a feature measured in _years_ (e.g. `YearRemodAdd`) consider shifting the data first. For example, you might subtract 1900 or 1910 from the year, so that a 1% increase in year is closer to meaning 1 year rather than 20 years.

</details>

```python
# Your code here - prepare data for modeling
features_to_transform = ['LotArea', 'MasVnrArea', 'GrLivArea']

# Apply log transformation to the features (adding 1 to avoid log(0) issues)
X_log_transformed = np.log(ames[features_to_transform] + 1)

# Apply log transformation to the target variable (SalePrice)
y_log_transformed = np.log(ames['SalePrice'] + 1)

# The X dataframe will contain the log-transformed features and the y series will be the log-transformed target
X_log_transformed.head(), y_log_transformed.head()
```

### Modeling

Now build a StatsModels OLS model with a log-transformed target as well as log-transformed features.

```python
# Your code here - build a model
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

# Add a constant column for the intercept term in the regression model
X_log_transformed_with_intercept = sm.add_constant(X_log_transformed)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_log_transformed_with_intercept, y_log_transformed, test_size=0.2, random_state=42)

# Fit the linear regression model using statsmodels
model_log = sm.OLS(y_train, X_train).fit()

# Print the model summary to see coefficients, R-squared, and p-values
print(model_log.summary())
```

### Model Evaluation and Interpretation

How did the model perform? How might we interpret its coefficients? Create as many cells as needed.

```python
# Your code here - evaluate the model
from sklearn.metrics import mean_squared_error, r2_score

# Predict on the test set
y_pred = model_log.predict(X_test)

# Calculate R-squared and MSE
r_squared = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

# Print model evaluation results
print(f"Model Evaluation:")
print(f"R-squared: {r_squared:.3f}")
print(f"Mean Squared Error: {mse:.2f}")

# Coefficients interpretation
print("\nCoefficients Interpretation:")
for feature, coef in model_log.params.items():
    if feature != 'const':  # Skip intercept
        print(f"{feature}: {coef:.4f} (for a 1% increase in {feature}, the price changes by {coef:.2f}% on average)")
```
